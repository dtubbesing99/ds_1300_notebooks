```python
serves = pd.read_csv('/users/dtubbesing/ds_1300_notebooks/work/serves.csv')
serves
```

```python
serv=serves.groupby(by='Nationality').mean()
ser=serv[['% 1st Serve','% 1st Serve Points Won','% 2nd Serve Points Won','% Service Games Won']]
ser
```

```python
s=ser.plot.bar(title='Serve Percentages by Nationality', figsize=(20,5))
s.set_ylabel('Serve Percentages')
```
